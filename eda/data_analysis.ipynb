```python
# Filtre les lignes où 'speaker' n'est pas null
df_preprocessed = df[df['speaker'].notna()]

# Compte les occurrences de chaque valeur non vide dans 'link_speaker'
value_counts = df_preprocessed['link_speaker'].value_counts()

# Nombre de lignes avec link_speaker non vide
nb_link_speaker_non_vide = df_preprocessed['link_speaker'].notna().sum()

# Ratio par rapport à df_preprocessed
ratio_link_speaker = (nb_link_speaker_non_vide / len(df_preprocessed)) * 100

print(f"Nombre de lignes avec link_speaker non vide : {nb_link_speaker_non_vide}")
print(f"Ratio par rapport à df_preprocessed : {ratio_link_speaker:.2f}%")
```

```python
df_preprocessed = df_preprocessed.copy()

df_preprocessed.loc[:, 'mp_id'] = (
    'PA' + df_preprocessed['link_speaker']
    .astype('string')
    .str.extract(r'/(\d+)\.asp(?:\?.*)?$', expand=False)
)
```

```python
# Lignes ayant un mp_id extrait depuis link_speaker
mask_mp_id_extrait = df_preprocessed['mp_id'].notna()

# Parmi ces lignes, vérifie si le mp_id existe dans politician
mask_bind_reussi = df_preprocessed.loc[mask_mp_id_extrait, 'mp_id'].isin(
    politician['mp_id'].dropna()
)

taux_bind = mask_bind_reussi.mean() * 100

print(f"Bind réussi parmi les mp_id extraits : {taux_bind:.3f} %")
print(f"Nombre de lignes testées : {mask_mp_id_extrait.sum():,}")
print(f"Nombre de lignes associées : {mask_bind_reussi.sum():,}")
```
